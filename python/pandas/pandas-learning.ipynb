```python
data = {
  "Duration":{
    "0":60,
    "1":60,
    "2":60,
    "3":45,
    "4":45,
    "5":60
  },
  "Pulse":{
    "0":110,
    "1":117,
    "2":103,
    "3":109,
    "4":117,
    "5":102
  },
  "Maxpulse":{
    "0":130,
    "1":145,
    "2":135,
    "3":175,
    "4":148,
    "5":127
  },
  "Calories":{
    "0":409,
    "1":479,
    "2":340,
    "3":282,
    "4":406,
    "5":300
  }
}

df = pd.DataFrame(data)
print(df)
```

```python
#dropping rows with empty cells
new_df = df.dropna()
print(new_df.to_string)
```

```python
#dropna(inplace = True) will NOT return a new DataFrame, but it will remove all rows containing NULL values from the original DataFrame
new_df = df.dropna(inplace = True)
print(new_df)
```

```python
#Replace Empty Values
new_df.fillna(100, inplace = True)
#Replace NULL values in the "Calories" columns with the number 130:
df["Calories"].fillna(130, inplace = True)
```

```python
#importing pandas and datas
result_data = {'name': ['Subramanian', 'Mallika', 'Karthika', 'Manju', 'Kousalya', 'Keerthana'],
              'score': [100, 100, 95, 95, 90, 85],
              'age': [50, 48, 29, 27, 25, 19],
              'qualify-label': ['yes', 'no', 'yes', 'no', 'yes', 'no']}

#creating dataframes
df = pd.DataFrame(result_data, index=None)

#computing number of rows
rows = len(df.axes[0])

#computing number of columns
columns = len(df.axes[1])

print(df)
print("Number of rows: ", rows)
print("Number of columns: ", columns)
```
